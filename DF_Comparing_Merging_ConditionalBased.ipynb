### The requirement is to merge both the dataframes and then if there is a data with same Item and same stock in both dataframes, we'll discard it else we have to consider that item. 

```python
import pandas as pd
import numpy as np
```

```python
woprice_df=pd.read_csv("teststock.csv",delimiter=';')
wprice_df=pd.read_csv("teststockreconprice.csv",delimiter=';')
```

```python
woprice_df
```

```python
wprice_df
```

```python
#Data of only wprice
dwprice_df=pd.merge(woprice_df,wprice_df,on='Item',how='left')
dwprice_df.fillna(0,inplace=True)
dwprice_df
```

```python
dwprice_df['Stock_y']=dwprice_df['Stock_y'].astype('int')
dwprice_df['Price']=dwprice_df['Price'].astype('int')
dwprice_df
```

```python
#Data of only woprice
dwoprice_df=pd.merge(woprice_df,wprice_df,on='Item',how='right')
dwoprice_df.fillna(0,inplace=True)
dwoprice_df
```

```python
dwoprice_df['Stock_x']=dwoprice_df['Stock_x'].astype('int')
dwoprice_df
```

```python
dwprice_df
```

```python
dwprice_df[((dwprice_df['Item']==dwprice_df['Item']) & (dwprice_df['Stock_x']==dwprice_df['Stock_y']))].index
```

```python
dwprice_df.drop(dwprice_df[((dwprice_df['Item']==dwprice_df['Item']) & (dwprice_df['Stock_x']==dwprice_df['Stock_y']))].index,inplace=True)
```

```python
dwoprice_df.drop(dwoprice_df[((dwoprice_df['Item']==dwoprice_df['Item']) & (dwoprice_df['Stock_x']==dwoprice_df['Stock_y']))].index,inplace=True)
```

```python
dwoprice_df
```

```python
dwprice_df
```

```python
final_df=pd.concat([dwoprice_df,dwprice_df]).reset_index()
final_df.drop(columns='index',inplace=True)
final_df.drop_duplicates(keep='first')
```

```python
final_df
```

```python
final_df['Stock_x']=abs(final_df['Stock_x']-final_df['Stock_y'])
final_df.rename(columns={"Stock_x":"Stock"},inplace=True)
final_df.drop(columns=['Stock_y'],axis=1,inplace=True)
final_df.drop_duplicates(keep='first',inplace=True)
final_df
```
